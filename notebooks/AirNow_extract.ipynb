```python
import requests
import csv
import time
import datetime
```

```python
def fetch_dates():
    starting_date = datetime.datetime(2025, 1, 1)
    length_of_time = 365
    dates = []
    for i in range(length_of_time):
        date = starting_date + datetime.timedelta(days=i)
        dates.append(date.strftime("%Y-%m-%d"))
    return dates
```

```python
key=None
class CurrentAirData:
    '''A class to fetch daily air quality data for zip codes in St. Louis.'''
    def __init__(self,key):

        self.AIRNOW_API_KEY = key
        self.AIRNOW_base_url = "https://www.airnowapi.org/aq/observation/zipCode/current/"

    
    def _get_city_location_data(self):
        with open("./zipcode_data/saint_louis_zipcodes.csv", newline="") as file:
            reader = csv.DictReader(
                file,
                fieldnames=["zip", "latitude", "longitude"]
            )
            return list(reader)

    def extract_air_quality_data(self):
        locations = self._get_city_location_data()
        pollution_data = []
        for row in locations:
            zip_code = row["zip"]
        
            params = {
                'zipCode': zip_code,
                'format' : 'application/json',
                'api_key': self.AIRNOW_API_KEY
            }
            try:
                response = requests.get(self.AIRNOW_base_url, params=params)
                time.sleep(2)
                pollution_data.append(response.json())
                print(f"Successfully requested data for zipcode: {zip_code}")
            except requests.exceptions.RequestException as e:
                print(f"Could not fetch data for {zip_code}. Error: {e}")
                break
    
    def extract_temperature_data(self):
        pass
    
```
