# What needs to happen:
## For each pollutant
* lags 1, 3, 7
* rolling 7 mean
* wind speed rolling 3
* cross feature: 1, 3


```python
data["NO2_r7"]=data['NO2'].shift(1).rolling(7).mean()
data["PM25_r7"]=data["PM25"].shift(1).rolling(7).mean()

data["NO2_l1"]=data['NO2'].shift(1)
data["NO2_l3"]=data['NO2'].shift(3)
data["NO2_l7"]=data['NO2'].shift(7)

data["PM25_l1"]=data['PM25'].shift(1)
data["PM25_l3"]=data['PM25'].shift(3)
data["PM25_l7"]=data['PM25'].shift(7)

data['wind_r3']=data['wind_speed_10m_max'].shift(1).rolling(3).mean()
```

```python
data = data.iloc[7:]
```

```python
data.to_csv("training_data.csv")
```
