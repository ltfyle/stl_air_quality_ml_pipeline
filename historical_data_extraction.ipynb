```python
KEY_AQS="mauvefox81"
EMAIL="ltfy4d@umsystem.edu"
def get_air_response(param_codes,county_code,key,email):
    """returns a response in json format for given parameters *up to five
    and given county code"""
    url_hist_air_quality="https://aqs.epa.gov/data/api/sampleData/byCounty?"
    KEY=key
    EMAIL=email

    params={
        'email': EMAIL,
        'key' : KEY,
        'param': param_codes,
        'bdate' : '20250101',
        'edate' : '20251231',
        'state' : '29',
        'county' : county_code
    }

    response=requests.get(url_hist_air_quality,params=params)
    if response.status_code==200:
        print("Success")
        return response.json()
    else:
        print(response.text)
```

```python
# codes 189 county, 510 city
def get_air_data(code):
    # lists of parameter codes
    # carbon monoxide, sulfur dioxide, nitrogen dioxide
    CO_SO2_NO2=['42101','42401','42602']
    Ozone_PM10_PM25=['44201','81102','88101']

    # responses for given county code, for params
    response_CO_SO2_NO2=get_air_response(CO_SO2_NO2,code)
    time.sleep(10)
    response_Ozone_PM10_PM25 = get_air_response(Ozone_PM10_PM25,code)

    # convert the json into pandas dataframes
    df_CO_SO2_NO2 = pd.json_normalize(response_CO_SO2_NO2, record_path=['Data'])
    df_Ozone_PM10_PM25=pd.json_normalize(response_Ozone_PM10_PM25, record_path=['Data'])
    # concatinate the dataframes 
    # for each set of pollutants into one dataframe
    air_quality=pd.concat([df_Ozone_PM10_PM25,df_CO_SO2_NO2], ignore_index=True)

    return air_quality
```

```python
# data dictionary
class DataDict:

    def __init__(self,df):
        """Collects all methods for creating a data dictionary"""
        self.df=df
        self.data_dict= pd.DataFrame({
        "column": self.df.columns,
        "dtype" : self.df.dtypes.astype(str).values,
        "missingness" : self.df.isna().mean()*100,
        "description": "",
        "notes" : ""
        })

    def get_data_dict(self):
        return self.data_dict
        
    def _check_initialized(self):
        if self.data_dict is None:
            raise ValueError("Data dictionary not created yet. Run set_data_dict() first.")
            

    def update_description(self,column,description):
        """Updates the description column"""
        self._check_initialized()

        mask=self.data_dict["column"]==column

        if not mask.any():
            print("column not found")
            return
        self.data_dict.loc[mask,"description"]=description
        print(self.data_dict.loc[mask])
        
    def update_notes(self,column,note):
        """Updates the notes section"""
        self._check_initialized()

        mask=self.data_dict["column"]==column
        if not mask.any():
            print("column not found")
            return
        self.data_dict.loc[mask,"notes"]=note
        print(self.data_dict.loc[mask])
    def remove_column(self, column):
        """Removes a column row from the data dictionary only"""
        self._check_initialized()

        self.data_dict = self.data_dict[self.data_dict["column"] != column]
```
