```python
import requests
```

```python
base = "https://api.weather.gov/openapi.json"

spec = requests.get(base).json()
```

```python
# first find out what the defintion contains
print(spec.keys())
```

```python
# get endpoints
paths = spec["paths"]

for path in paths.keys():
    print(path)

    methods = paths[path]
    for method in methods.keys():
        print(" ", method)
```

```python
# an example of an endpoint
endpoint = "/alerts/active/zone/{zoneId}"

for key, value in spec['paths'][endpoint].items():
    print(key)
    print(value)
```

```python
endpoint = "/products"
# endpoint -> method -> details -> fields
'''
endpont: products
method: get
details: contains fields

paths
 └── /products
      └── get
           ├── parameters   (list)
           ├── responses    (dict)
           ├── description  (string)
'''
for method, details in spec['paths'][endpoint].items():

    print("\nMethod:\n",method)
    for field, value in details.items():
        print("Field\n", field)

              # If it's a list (like parameters), print it more clearly
        if isinstance(value, list):
            for item in value:
                print("    -", item)

        # If it's a dict (like responses), print keys
        elif isinstance(value, dict):
            for subkey in value:
                print("    -", subkey)

        # Otherwise print the raw value (description, etc.)
        else:
            print("    ", value)
```
